```python
def choose_teacher_out(l,s):
    len_of_s = len(s)#计算字符串的长度
    m = []
    for i in l:
        if(i[:len_of_s]==s):
            m.append(i)
    return(m)
```

```python
hxqk = choose_teacher_out(l,'http://navi.cnki.net/KNavi/pubDetail?pubtype=journal&pcode=CJFD')
```

```python
driver.find_element_by_link_text('下一页').click()
l = [i.get_attribute('href') for i in driver.find_elements_by_tag_name('a')]
l = list(map(lambda x:str(x),l))
hxqk.extend(choose_teacher_out(l,'http://navi.cnki.net/KNavi/pubDetail?pubtype=journal&pcode=CJFD'))
```

```python
hxqk
```

```python
message_all_hxqk = []
for each_url_hxqk in hxqk:
    each_message = []
    driver.get(each_url_hxqk)
    time.sleep(3)
    #期刊名称
    each_message.append(driver.find_element_by_xpath("//h3[contains(@class,'titbox')]").text)
    ###模拟点击栏目浏览
    driver.find_element_by_link_text('栏目浏览').click()
    time.sleep(3)
    l = [i.text for i in driver.find_element_by_xpath("//div[contains(@class,'page-list')]").find_elements_by_tag_name('a')]
    try:
        if(l==[]):each_message.append(driver.find_element_by_id('collayer+0').text)
        else:
            count = int([i.text for i in driver.find_element_by_xpath("//div[contains(@class,'page-list')]").find_elements_by_tag_name('a')][-2])
            each_message.append(driver.find_element_by_id('collayer+0').text)
            for i in range(count-1):
                driver.find_element_by_class_name('next').click()
                time.sleep(3)
                each_message.append(driver.find_element_by_id('collayer+%s'%str(i+1)).text)
    except:
        continue
    message_all_hxqk.append(each_message)
```
